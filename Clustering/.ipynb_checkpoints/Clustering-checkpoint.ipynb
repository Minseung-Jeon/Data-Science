### Exercise 1

Using the Python “sklearn” library, create and visualize the k-means clusters (with k=5) for the given heart disease dataset. For visualization, draw the scatter plot using the age and cholesterol features on each group of clusters.

```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from kneed import KneeLocator
import warnings
warnings.filterwarnings('ignore')

# load dataset
data = pd.read_csv("processed.cleveland.data")
for col in data.columns:
    data = data[data[col] != '?']
X = data.iloc[:,:].values

# running k means algorithm
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
data['cluster'] = y_kmeans

# visualization of clusters
plt.scatter(X[y_kmeans == 0,0], X[y_kmeans == 0,4], c = 'red', label = '0')
plt.scatter(X[y_kmeans == 1,0], X[y_kmeans == 1,4], c = 'blue', label = '1')
plt.scatter(X[y_kmeans == 2,0], X[y_kmeans == 2,4], c = 'green', label = '2')
plt.scatter(X[y_kmeans == 3,0], X[y_kmeans == 3,4], c = 'orange', label = '3')
plt.scatter(X[y_kmeans == 4,0], X[y_kmeans == 4,4], c = 'purple', label = '4')

plt.title('K-Means Clusters Visualization')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend()
plt.show()

elbow_locator = KneeLocator(range(1,11), sse, curve = 'convex', direction = 'decreasing')
k_estimation = elbow_locator.elbow
print(k_estimation)


```

### Exercise 2

Apply k-means clusters on the heart disease dataset with varying numbers of clusters from 1 to 10 and compute their corresponding Sum of squared Error (SSE) value. Plot the graph using Python “matplotlib” library and estimate the right “k” value.

```python
# array to store sse values
sse = []
# calculate sse values for num of clusters ranging 1-10
for num in range(1,11):
    k_means = KMeans(n_clusters = num)
    k_means.fit(X)
    sse.append(k_means.inertia_)
    print(f'The SSE for {num} cluster(s) is {k_means.inertia_}')

# plot sse vs num of clusters graph
plt.figure()
plt.title('SSE vs Num of Clusters for K-Means')
plt.xlabel('# of Clusters')
plt.ylabel('SSE (Sum of Squared Error)')
plt.plot(range(1,11), sse)
plt.show()
```

```python
# find the elbow point on the graph to get optimal K
kl = KneeLocator(range(1,11), sse, curve = 'convex', direction = 'decreasing')
kl.plot_knee()
print('We can estimate the elbow point to be 3.')
```

### Exercise 3

Create and visualize the k-means clustering with the “k” value obtained in Q2. The clustering algorithm ultimately groups similar patients by matching its features. Thus, for the visualization, draw the scatter plot using the age and cholesterol features on each cluster group and label them as “Group-A”, “Group-B” etc.,

```python
# running k means algorithm with optimal k
optimal_k = 3
optimal_kmeans = KMeans(n_clusters = optimal_k, init = 'k-means++', random_state = 42)
y_optimal_kmeans = optimal_kmeans.fit_predict(X)
data['cluster'] = y_optimal_kmeans

# visualizing the clusters
plt.scatter(X[y_optimal_kmeans == 0,0], X[y_optimal_kmeans == 0,4], c = 'red', label = 'Group-A')
plt.scatter(X[y_optimal_kmeans == 1,0], X[y_optimal_kmeans == 1,4], c = 'blue', label = 'Group-B')
plt.scatter(X[y_optimal_kmeans == 2,0], X[y_optimal_kmeans == 2,4], c = 'green', label = 'Group-C')
plt.title('K-Means Clusters Visualization')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend()
plt.show()

```

### Exercise 4

Plot a dendrogram using Python scipy.cluster.hierarchy method.

```python
import scipy.cluster.hierarchy as sch

# plotting dendrogram
dendrogram = sch.dendrogram(sch.linkage(X , method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('X')
plt.ylabel('Euclidean Distance')
plt.show()
```

### Exercise 5

Create the agglomerative clustering with the number of clusters is equal to the “k” value obtained in Question-2. Visualize the clusters similar to Question-2.

```python
from sklearn.cluster import AgglomerativeClustering

# running agglomerative clustering algorithm
hc = AgglomerativeClustering(n_clusters = optimal_k, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)

# visualizing clusters
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,4], c = 'red', label = 'Group-A')
plt.scatter(X[y_hc == 1,0], X[y_hc == 1,4], c = 'blue', label = 'Group-B')
plt.scatter(X[y_hc == 2,0], X[y_hc == 2,4], c = 'green', label = 'Group-C')
plt.title('Agglomerative Clustering Visualization')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend()
plt.show()
```

### Exercise 6

Compute silhouette score for both K-means and agglomerative clustering and tell us which clustering is better for the given dataset.

```python
from sklearn.metrics import silhouette_score

# calculating silhouette score for each clustering method
kmeans_score = silhouette_score(X, optimal_kmeans.labels_)
agglomerative_score = silhouette_score(X, hc.labels_)

# output scores
print(f'K-Means Score: {kmeans_score}')
print(f'Agglomerative Score: {agglomerative_score}')

# output which one is better
if kmeans_score > agglomerative_score:
    print('\nK-Means is better')
elif kmeans_score < agglomerative_score:
    print('\nAgglomerative is better')
else:
    print('\nThey are the same')
```
