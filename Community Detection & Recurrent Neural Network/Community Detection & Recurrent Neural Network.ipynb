# Girvan Newman Algorithm

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community

# creating graph Object
G = nx.Graph()
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'H'), ('H', 'I'), ('H', 'G'), ('I', 'G'), ('G', 'E'), ('D', 'E'), ('E', 'F'), ('D', 'F'), ('C', 'D')]
G.add_edges_from(edges)

# calculating the edge betweenness
edge_betweenness = nx.edge_betweenness_centrality(G).items()
edge_betweenness = sorted(edge_betweenness, key=lambda pair: -pair[1])
print(edge_betweenness)

#Draw original Graph
nx.draw(G, with_labels=True, node_color='r')
plt.title("Original Graph")
plt.show()


step = 0
betweenness_value = 100
threshold = 0.26
while betweenness_value > threshold:
    step=step+1
    # calculate edge betweenness and converting from dictionary to list
    edge_betweenness = nx.edge_betweenness_centrality(G).items()
    
    # sort edge betweenness (sort second element of each pair)
    # we get highest betweenness on top and lowest betweenness on bottom
    # delete the highest betweenness
    edge_to_delete = sorted(edge_betweenness, key=lambda pair: -pair[1])[0][0]
    G.remove_edge(*edge_to_delete)
    
    #get betweenness value
    betweenness_value = sorted(edge_betweenness, key=lambda pair: -pair[1])[0][1]
    
    #draw graph by each step until threshold (proper number) 
    nx.draw(G, with_labels=True, node_color='r')
    plt.title("step: %s\nEdge %s Deleted\nbetweenness_value is %s"%(step, edge_to_delete, betweenness_value))
    plt.show()
```

# Louvain Algorithm

```python
import warnings
warnings.filterwarnings("ignore")

import networkx as nx
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import community as community_louvain

# creating graph Object
G = nx.Graph()
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'H'), ('H', 'I'), ('H', 'G'), ('I', 'G'), ('G', 'E'), ('D', 'E'), ('E', 'F'), ('D', 'F'), ('C', 'D')]
G.add_edges_from(edges)

#run louvain algorithm
partition = community_louvain.best_partition(G)

#draw graph
pos = nx.spring_layout(G)

# color the nodes according to their partition
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=40,
                       cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)

plt.title("Louvain Algorithm")
plt.show()

```

# Greedy Modularity Communities Algorithm

```python
import warnings
warnings.filterwarnings("ignore")

import networkx as nx
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from networkx.algorithms import community 

# creating graph Object
G = nx.Graph()
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'H'), ('H', 'I'), ('H', 'G'), ('I', 'G'), ('G', 'E'), ('D', 'E'), ('E', 'F'), ('D', 'F'), ('C', 'D')]
G.add_edges_from(edges)

# Find communities using greedy modularity
communities = community.greedy_modularity_communities(G)


# Print the communities
for i, community_nodes in enumerate(communities):
    print(f"Community {i}: {community_nodes}")

# Visualization
pos = nx.spring_layout(G)
cmap = cm.get_cmap('viridis', len(communities))  # Adjust colormap for number of communities

# Draw nodes, colored by community
for i, community_nodes in enumerate(communities):
    nx.draw_networkx_nodes(G, pos, nodelist=community_nodes,
                           node_size=40, cmap=cmap, node_color=cmap(i))

nx.draw_networkx_edges(G, pos, alpha=0.5)

plt.title("Greedy Modularity Communities Algorithm")
plt.show()

```
